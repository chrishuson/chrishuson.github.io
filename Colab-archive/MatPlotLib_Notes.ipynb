```python
# Bar chart, stacked, error bars, shaded
data1 = [2, 3, 5, 6, 4]
data2 = [4, 3, 3, 1, 3]

plt.subplot(1,3,1)
plt.bar(range(5), data1)
plt.bar(range(5), data2, bottom=data1)

plt.subplot(1,3,2)
plt.bar(range(len(data1)), data1, yerr=2, capsize=5) # yerr also accepts list or pair of lists

plt.subplot(1,3,3)
y_lower = [0.9 * y for y in data1]
y_upper = [1.1 * y for y in data1]
plt.fill_between(range(5), y_lower, y_upper, alpha=0.2)
plt.plot(data1)

plt.show()
```

```python
cc = pd.read_csv('Codecademy/roller_coaster_starting/roller_coasters.csv') #captaincoasters.com

def plot_condition(condition, df):
    counts = df.groupby('status').name.count()[condition]
    plt.pie(counts, labels=condition, autopct='%d%%')
    plt.show()
    
plot_condition(['status.operating','status.closed.definitely'], cc)
```

```python
def plot_inversions(park, df):
    park_df = df[df.park == park]
    plt.figure(figsize=(9,4))
    ax = plt.subplot()
    plt.bar(range(len(park_df)), park_df.num_inversions)
    plt.title('Number of inversions in each coaster\nat the park ' + park)
    ax.set_xticks(range(len(park_df)))
    ax.set_xticklabels(park_df.name, rotation=80)
    plt.show()
    return park_df

plot_inversions('Six Flags Magic Mountain', cc)
```
